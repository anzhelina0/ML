```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
```

```python
class ConvNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(3, 16, kernel_size = 3, padding = 1) # 16, 32, 32
        self.relu1 = nn.ReLU()
    
        self.conv2 = nn.Conv2d(16, 32, kernel_size = 3, padding = 1) # 32, 32, 32
        self.relu2 = nn.ReLU()
        self.pool1 = nn.MaxPool2d(2, 2) # 32, 16, 16
    
        self.conv3 = nn.Conv2d(32, 64, kernel_size = 3, padding = 1) # 64, 16, 16
        self.relu3 = nn.ReLU()
    
        self.conv4 = nn.Conv2d(64, 64, kernel_size = 3, padding = 1) # 64, 16, 16
        self.relu4 = nn.ReLU()
    
        self.conv5 = nn.Conv2d(64, 64, kernel_size = 3, padding = 1) # 64, 16, 16
        self.relu5 = nn.ReLU()
        self.pool2 = nn.MaxPool2d(2, 2) # 64, 8, 8
    
        self.conv6 = nn.Conv2d(64, 64, kernel_size = 3, padding = 1) # 64, 8, 8
        self.relu6 = nn.ReLU()

    def forward(self, x):
        x = self.conv1(x)
        x = self.relu1(x)

        x = self.conv2(x)
        x = self.relu2(x)
        x = self.pool1(x)

        x = self.conv3(x)
        x = self.relu3(x)

        x = self.conv4(x)
        x = self.relu4(x)

        x = self.conv5(x)
        x = self.relu5(x)
        x = self.pool2(x)

        x = self.conv6(x)
        x = self.relu6(x)

        return x
```

```python
model = ConvNet()
```

```python
test_tensor = torch.randn(1, 3, 32, 32)
```

```python
output = model(test_tensor)
```

```python
print(output.shape)
```

```python
criterion = nn.MSELoss()
target = torch.randn_like(output)
```

```python
optimizer = torch.optim.Adam(model.parameters(), lr = 0.001)
```

```python
optimizer.zero_grad()
```

```python
loss = criterion(output, target)
```

```python
loss.backward()
```

```python
print("=== ГРАДИЕНТЫ ПО СЛОЯМ ===")
for name, param in model.named_parameters():
    if param.grad is not None:
        grad_mean = param.grad.mean().item()
        print(f"{name}: {grad_mean:.8f}")
```
